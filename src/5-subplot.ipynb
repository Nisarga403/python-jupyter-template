### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y_line = x**2
y_scatter = 2 * x
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y_line)
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('x^2')
axs[0, 0].grid(True)

# Bar plot
colors = ['red', 'blue', 'green', 'purple']
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Chart')

# Scatter plot
axs[1, 0].scatter(x, y_scatter)
for i in range(len(x)):
    axs[1, 0].annotate(f'({x[i]}, {y_scatter[i]})', (x[i], y_scatter[i]))
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('2 * x')

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
axs[1, 1].axis('equal')  # Ensure that pie chart is a circle

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```
